```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python
# Load the datasets
restaurants_df = pd.read_csv("../data/restaurants_cleaned.csv")
menus_df = pd.read_csv("../data/cleaned_restaurant_menus.csv")
cuisines_df = pd.read_csv("../data/restaurant_cuisines.csv")
```

```python
# Set UberEats color palette
ubereats_color = "#06C167"  # UberEats green
sns.set_palette([ubereats_color])
```

```python
# Distribution of Restaurant Ratings
plt.figure(figsize=(10, 5))
sns.histplot(restaurants_df['score'], bins=20, kde=True)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 Restaurant Categories by Frequency
plt.figure(figsize=(12, 6))
categories_counts = restaurants_df['category'].value_counts().head(10)
sns.barplot(x=categories_counts.values, y=categories_counts.index)
plt.title('Top 10 Restaurant Categories')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

```python
# Most Common Cuisines
plt.figure(figsize=(12, 6))
cuisine_counts = cuisines_df['cuisine_type'].value_counts().head(10)
sns.barplot(x=cuisine_counts.values, y=cuisine_counts.index)
plt.title('Top 10 Most Common Cuisines')
plt.xlabel('Count')
plt.ylabel('Cuisine Type')
plt.show()
```

```python
# Fix Price Distribution of Menu Items (Histogram with Binned Prices)
bins = [0, 5, 10, 15, 20, 30, 50, 100]
plt.figure(figsize=(10, 5))
sns.histplot(menus_df[(menus_df['price'] > 0) & (menus_df['price'] < 100)]['price'], bins=bins, color=ubereats_color, kde=True)
plt.title('Distribution of Menu Prices (Binned)')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.xticks(bins)
plt.show()
```

```python
# Box Plot: Price Distribution Across Menu Categories
plt.figure(figsize=(12, 6))
sns.boxplot(data=menus_df[(menus_df['price'] > 0) & (menus_df['price'] < 100)], x='category', y='price', color=ubereats_color)
plt.xticks(rotation=45, ha='right')
plt.title('Price Distribution Across Menu Categories')
plt.xlabel('Menu Category')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Fix Scatter plot of Ratings vs Number of Reviews (Reduce Overcrowding)
filtered_reviews = restaurants_df[restaurants_df['ratings'] <= restaurants_df['ratings'].quantile(0.95)]
plt.figure(figsize=(10, 6))
sns.scatterplot(x=filtered_reviews['ratings'], y=filtered_reviews['score'], alpha=0.3, color=ubereats_color)
plt.title('Ratings vs. Number of Reviews (Filtered)')
plt.xlabel('Number of Reviews')
plt.ylabel('Score')
plt.show()
```

```python
# Fix Scatter plot of Ratings vs Number of Reviews (Hexbin + Transparency)
filtered_reviews = restaurants_df[restaurants_df['ratings'] <= restaurants_df['ratings'].quantile(0.95)]
plt.figure(figsize=(10, 6))
plt.hexbin(filtered_reviews['ratings'], filtered_reviews['score'], gridsize=50, cmap='Greens', mincnt=1)
plt.colorbar(label='Count of Restaurants')
plt.title('Ratings vs. Number of Reviews (Hexbin)')
plt.xlabel('Number of Reviews')
plt.ylabel('Score')
plt.show()
```

```python
# Alternative: 2D KDE Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=filtered_reviews, x='ratings', y='score', fill=True, cmap='Greens', levels=50)
plt.title('Ratings vs. Number of Reviews (Density)')
plt.xlabel('Number of Reviews')
plt.ylabel('Score')
plt.show()
```

```python
# Merge price data with restaurants
restaurants_with_price = restaurants_df.merge(
    menus_df.groupby('restaurant_id')['price'].mean().reset_index(), 
    left_on='id', right_on='restaurant_id', how='left'
)
```

```python
# Hypothesis 1: Correlation between price and restaurant rating
price_rating_corr = restaurants_with_price[['score', 'price']].corr().iloc[0,1]
print(f"Correlation between price and rating: {price_rating_corr:.2f}")
```

```python
# Hypothesis 2: Average rating per cuisine type
cuisine_ratings = cuisines_df.merge(
    restaurants_df[['id', 'score']], left_on='restaurant_id', right_on='id', how='left'
)
cuisine_avg_ratings = cuisine_ratings.groupby('cuisine_type')['score'].mean().reset_index().sort_values(by='score', ascending=False)
print("\nTop 10 Highest-Rated Cuisines:")
print(cuisine_avg_ratings.head(10))
```

```python
# Perform statistical test: Do cuisine ratings significantly differ?
cuisine_groups = [group['score'].dropna().values for _, group in cuisine_ratings.groupby('cuisine_type')]
anova_result = stats.f_oneway(*cuisine_groups)
print(f"\nANOVA test p-value: {anova_result.pvalue:.5f}")
```

```python
# Box Plot: Distribution of Ratings Across Different Cuisines
plt.figure(figsize=(14, 6))
top_cuisines = cuisine_ratings['cuisine_type'].value_counts().head(15).index  # Select top 15 cuisines for readability
filtered_cuisines = cuisine_ratings[cuisine_ratings['cuisine_type'].isin(top_cuisines)]

sns.boxplot(data=filtered_cuisines, x='cuisine_type', y='score', color=ubereats_color)
plt.xticks(rotation=45, ha='right')
plt.title('Distribution of Ratings Across Top 15 Cuisines')
plt.xlabel('Cuisine Type')
plt.ylabel('Rating')
plt.show()
```

```python
# Bar Chart: Average Rating per Cuisine Type
plt.figure(figsize=(14, 6))
sns.barplot(data=cuisine_avg_ratings.head(15), x='score', y='cuisine_type', color=ubereats_color)
plt.title('Average Rating of Top 15 Cuisines')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine Type')
plt.xlim(3, 5)
plt.show()
```
