# Load libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data

```python
path = "../data/raw/restaurants.csv"

restaurants_df = pd.read_csv(path)
```

# Display dataset information

```python
restaurants_df.info()
restaurants_df.head()
```

# Check missing values

```python
missing_values = restaurants_df.isnull().sum()
print("Missing Values:\n", missing_values)

# Fix inplace method issue by assigning the result back to the DataFrame
restaurants_df = restaurants_df.assign(
    score=restaurants_df["score"].fillna(0),
    ratings=restaurants_df["ratings"].fillna(0),
    category=restaurants_df["category"].fillna("Unknown"),
    price_range=restaurants_df["price_range"].fillna("Not Available"),
    full_address=restaurants_df["full_address"].fillna("Unknown"),
    zip_code=restaurants_df["zip_code"].fillna("Unknown"),
)

# Verify that missing values are handled
print("\nMissing Values After Cleaning:\n", restaurants_df.isnull().sum())

# Display dataset after cleaning
restaurants_df.head()
```

# 5️⃣ Fix `price_range` column (REMOVE the incorrect `$` → `Dollar` replacement)

```python
# 5️⃣ Fix `price_range` column (REMOVE the incorrect `$` → `Dollar` replacement)
price_mapping = {
    "$": "Cheap",
    "$$": "Moderate",
    "$$$": "Expensive",
    "$$$$": "Very Expensive"
}

restaurants_df["price_range"] = restaurants_df["price_range"].apply(lambda x: price_mapping[x] if x in price_mapping else x)

# 6️⃣ Print unique values to confirm mapping worked
print("Unique Price Ranges After Fix:", restaurants_df["price_range"].unique())

# 7️⃣ Set plot style
sns.set_style("whitegrid")

# 8️⃣ Distribution of Restaurant Scores
plt.figure(figsize=(8,5))
sns.histplot(restaurants_df["score"], bins=20, kde=True)
plt.title("Distribution of Restaurant Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()

# 9️⃣ Corrected Count of Restaurants by Price Range
# Remove unexpected values that didn't get mapped correctly
valid_price_ranges = ["Cheap", "Moderate", "Expensive", "Very Expensive", "Not Available"]
restaurants_df = restaurants_df[restaurants_df["price_range"].isin(valid_price_ranges)]

plt.figure(figsize=(8,5))
sns.countplot(y=restaurants_df["price_range"], order=restaurants_df["price_range"].value_counts().index)
plt.title("Count of Restaurants by Price Range (Corrected)")
plt.xlabel("Count")
plt.ylabel("Price Range")
plt.xticks(rotation=45)
plt.show()

# 🔟 Top 10 Most Common Restaurant Categories
plt.figure(figsize=(10,5))
top_categories = restaurants_df["category"].value_counts().nlargest(10)
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Most Common Restaurant Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
# Get unique restaurant categories
unique_categories = restaurants_df["category"].dropna().unique()

# Display all unique categories (limited to first 100 for readability)
print("🔍 Unique Restaurant Categories (First 100):")
print(unique_categories[:100])  # Change the number if you want to see more

# Count occurrences of each category
category_counts = restaurants_df["category"].value_counts()

# Display the top 20 most common categories
print("\n📊 Top 20 Most Frequent Categories:")
print(category_counts.head(20))
```

```python
# Split the 'category' column into individual cuisine types and count occurrences
category_counts = restaurants_df['category'].dropna().str.split(', ').explode().value_counts()

# Display the results
print(category_counts)
```

```python
# Create a new table for restaurant_cuisines by splitting categories into individual cuisine types
restaurant_cuisines = restaurants_df[['id', 'category']].dropna().copy()
restaurant_cuisines['category'] = restaurant_cuisines['category'].str.split(', ')
restaurant_cuisines = restaurant_cuisines.explode('category')

# Rename columns for clarity
restaurant_cuisines.rename(columns={'id': 'restaurant_id', 'category': 'cuisine_type'}, inplace=True)

# Save both cleaned datasets
restaurants_cleaned_path = "../data/clean/restaurants_cleaned.csv"  # Replace with desired path
restaurant_cuisines_path = "../data/clean/restaurant_cuisines.csv"  # Replace with desired path

restaurants_df.to_csv(restaurants_cleaned_path, index=False)
restaurant_cuisines.to_csv(restaurant_cuisines_path, index=False)

# Display confirmation message
print("Cleaned datasets saved successfully.")
```
